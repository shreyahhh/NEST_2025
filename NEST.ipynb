```python
for col in ['Allocation', 'Intervention Model', 'Masking', 'Primary Purpose']:
    df[col] = df[col].str.strip()

#Remove the prefixes from the columns
prefixes_to_remove = ["Allocation: ", "Intervention Model: ", "Masking: ", "Primary Purpose: "]
for col, prefix in zip(['Allocation', 'Intervention Model', 'Masking', 'Primary Purpose'], prefixes_to_remove):
    df[col] = df[col].str.replace(prefix, '', regex=False)

df
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = ["Allocation", "Intervention Model", "Masking", "Primary Purpose"]

# Create indexed column names for these columns
categorical_columns_indexed = [col + "_Indexed" for col in categorical_columns]

# Apply LabelEncoder to each specified column
label_encoders = {}
for col, indexed_col in zip(categorical_columns, categorical_columns_indexed):
    le = LabelEncoder()
    df[indexed_col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Save the encoder for inverse transformation if needed

df
```

```python
# Create DataFrame
df = pd.DataFrame(data)

def get_unique_values(column):
    unique_values = set()
    for entry in df[column].dropna():
        unique_values.update(map(str.strip, str(entry).replace(",", "|").split("|")))
    return sorted(unique_values)  # Sort for consistent encoding

# Extract unique values for each column
unique_age = get_unique_values("Age")
unique_sex = get_unique_values("Sex")
unique_phases = get_unique_values("Phases")

# Function to encode a single column based on unique values
def encode_column(row, unique_values, column_name):
    values = set(map(str.strip, str(row[column_name]).replace(",", "|").split("|")))
    return ''.join(['1' if val in values else '0' for val in unique_values])

# Apply encoding to each column and create separate encoded columns
df["Age_Encoded"] = df.apply(lambda row: encode_column(row, unique_age, "Age"), axis=1)
df["Sex_Encoded"] = df.apply(lambda row: encode_column(row, unique_sex, "Sex"), axis=1)
df["Phases_Encoded"] = df.apply(lambda row: encode_column(row, unique_phases, "Phases"), axis=1)

# Output the resulting DataFrame
print(df)
```

```python
df.to_csv("final_dataset.csv", index = False)
```
