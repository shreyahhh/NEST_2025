```python
import pandas as pd
df= pd.read_csv("final_dataset_1.csv")
```

```python
df.columns
```

```python
df.drop(columns=["Allocation","Conditions", "Interventions", 'Sex', 'Age', 'Phases','Start Date',
       'Completion Date', 'Last Update Posted', 'Locations', 'Intervention Model','Masking', 'Primary Purpose'])
```

```python
df.columns
```

```python
df = df.drop(["Allocation","Conditions", "Interventions", 'Sex', 'Age', 'Phases','Start Date',
       'Completion Date', 'Last Update Posted', 'Locations', 'Intervention Model','Masking', 'Primary Purpose'], axis=1)
```

```python
df.columns
```

```python
df = df.drop(['Study Status', 'Study Results', 'Study Type', 'Study Design', 'Funder Type'], axis=1)
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder

# Calculate value counts
value_counts = df['Sponsor'].value_counts()

# Identify sponsors with frequency > 100
sponsors_to_keep = value_counts[value_counts > 50].index

# Encode sponsors
df['Sponsor_encoded'] = df['Sponsor'].apply(lambda x: x if x in sponsors_to_keep else 'Other')
label_encoder = LabelEncoder()
df['Sponsor_encoded'] = label_encoder.fit_transform(df['Sponsor_encoded'])
```

```python
df = df.drop(['Sponsor'], axis=1)
```
